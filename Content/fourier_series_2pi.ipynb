# Fourier series of simple functions with period $2\pi$

This notebook illustrates the Fourier series expansion for simple functions with period $2\pi$. Some examples can be found at:

https://www.math24.net/fourier-series-definition-typical-examples/

```python
import numpy as np
from scipy.signal import sawtooth, square
import matplotlib.pyplot as plt
import my_functions as mfun
```

### Upward sawtooth function

```python
nmax = 11
bn_up_sawtooth = mfun.upward_sawtooth_bn(nmax)
x = np.linspace(-10, 10, 101)

plt.figure(figsize=(10,6))
plt.plot(x, sawtooth(t=x-np.pi, width=1), 'k.', label='upward sawtooth')
for n in range(1, nmax+1, 4):
    fs = mfun.fourier_series_real(x=x, a0=0, an=None, bn=bn_up_sawtooth[:n])
    plt.plot(x, fs, '-', label='n = %d' % n)
plt.xticks(ticks=[i*np.pi for i in range(-4, 4)], 
           labels=[('%d $\pi$' % i) for i in range(-4, 4)])
plt.xlim(np.min(x), np.max(x))
plt.grid()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('x', fontsize=16)
plt.legend(loc='upper right', fontsize=12)
plt.show()
```

```python
nmax = 20
a0_up_sawtooth = 0
an_up_sawtooth = np.zeros(nmax)
bn_up_sawtooth = mfun.upward_sawtooth_bn(nmax)
x = np.linspace(-10, 10, 201)
fs_real_up_sawtooth = mfun.fourier_series_real(x=x, a0=a0_up_sawtooth, 
                                              an=an_up_sawtooth, bn=bn_up_sawtooth)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, sawtooth(t=x-np.pi, width=1), 'k.', label='up sawtooth')

plt.plot(x, fs_real_up_sawtooth, 'b-', label='fourier real')

plt.xticks(ticks=[i*np.pi for i in range(-4, 4)], 
           labels=[('%d $\pi$' % i) for i in range(-4, 4)])
plt.xlim(np.min(x), np.max(x))
plt.grid()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('x', fontsize=16)
plt.legend(loc='upper right', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title('upward sawtooth - nmax %d' % nmax, fontsize=16)
plt.plot(np.hstack([a0_up_sawtooth, an_up_sawtooth]), 'bo--', label='an')
plt.plot(np.hstack([             0, bn_up_sawtooth]), 'ro--', label='bn')
plt.xticks(ticks=[i for i in range(bn_up_sawtooth.size+1)])
plt.grid()
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(loc='upper right', fontsize=12)
plt.xlabel('n', fontsize=16)
plt.show()
```

```python
c0_up_sawtooth, cn_up_sawtooth = mfun.complex_coefficients(a0_up_sawtooth, 
                                                           an_up_sawtooth, 
                                                           bn_up_sawtooth)
```

```python
nmax = cn_up_sawtooth.size
x = np.linspace(-10, 10, 201)
fs_complex_up_sawtooth = mfun.fourier_series_complex(x=x, c0=c0_up_sawtooth, 
                                                     cn=cn_up_sawtooth)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, sawtooth(t=x-np.pi, width=1), 'k.', label='up sawtooth')

plt.plot(x, fs_complex_up_sawtooth.real, 'b-', label='fourier complex real')
plt.plot(x, fs_complex_up_sawtooth.imag, 'r-', label='fourier complex imag')

plt.xticks(ticks=[i*np.pi for i in range(-4, 4)], 
           labels=[('%d $\pi$' % i) for i in range(-4, 4)])
plt.xlim(np.min(x), np.max(x))
plt.grid()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('x', fontsize=16)
plt.legend(loc='upper right', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title('upward sawtooth - nmax %d' % nmax, fontsize=16)
plt.plot(np.hstack([cn_up_sawtooth[::-1].real,
                    c0_up_sawtooth, 
                    cn_up_sawtooth.real]), 'bo--', label='cn real')
plt.plot(np.hstack([cn_up_sawtooth[::-1].imag,
                    0, 
                    cn_up_sawtooth.imag]), 'ro--', label='cn imag')
plt.xticks(ticks=[i for i in range(2*cn_up_sawtooth.size+1)],
           labels=[i for i in range(-cn_up_sawtooth.size, cn_up_sawtooth.size+1)],
           rotation=70)
plt.grid()
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('n', fontsize=16)
plt.legend(loc='upper right', fontsize=12)
plt.show()
```

### Odd square function

```python
nmax = 11
bn_odd_square = mfun.odd_square_bn(nmax)
x = np.linspace(-10, 10, 101)

plt.figure(figsize=(10,6))
plt.plot(x, 0.5*(square(t=x, duty=0.5) + 1), 'k.', label='odd square')
for n in range(1, nmax+1, 4):
    fs_real_odd_square = mfun.fourier_series_real(x=x, a0=1, an=None, bn=bn_odd_square[:n])
    plt.plot(x, fs_real_odd_square, '-', label='n = %d' % n)
plt.xticks(ticks=[i*np.pi for i in range(-4, 4)], 
           labels=[('%d $\pi$' % i) for i in range(-4, 4)])
plt.xlim(np.min(x), np.max(x))
plt.grid()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('x', fontsize=16)
plt.legend(loc='upper right', fontsize=12)
plt.show()
```
